```python
import numpy as np
import matplotlib.pyplot as plt
import math	
from sympy import *
from matplotlib import cm

#  
A = 3; B = 4; C = -4; D = -12
f = lambda x, y: (-A * x - B * y - D) / C #  
fig = plt.figure(figsize=(10, 10)) #  
ax = fig.add_subplot(1, 1, 1, projection='3d') # 
xval = np.linspace(-4, 4, 100) #   x
yval = np.linspace(-4, 4, 100) #   y
x, y = np.meshgrid(xval, yval)
p = np.array([0, 0, - 3]) #   
v = np.array([3, 4, -4]) #  
ax.quiver(*p, *v, color='Blue') #   
z = f(x, y) #   z  
surf = ax.plot_surface(x, y, z, rstride=10, cstride=10, cmap=cm.plasma) #  
plt.xlabel('x')
plt.ylabel('y')
plt.title(' 3x + 4y  4z  12 = 0')
ax.view_init(10, -30)
plt.show()
```
